```python
geo_df = df.dropna(subset=["Latitude", "Longitude"]).copy()

geo_df = geo_df[
    (geo_df["Latitude"] != 0) &
    (geo_df["Longitude"] != 0)
]
```

```python
geo_df.shape
```

```python
import folium
from folium.plugins import FastMarkerCluster

restaurant_map = folium.Map(
    location=[
        geo_df["Latitude"].mean(),
        geo_df["Longitude"].mean()
    ],
    zoom_start=2
)

locations = geo_df[["Latitude", "Longitude"]].values.tolist()

FastMarkerCluster(locations).add_to(restaurant_map)

restaurant_map
```

```python
restaurant_map.save("restaurant_locations_map.html")
```

```python
location_rating_correlation = df[
    ["Latitude", "Longitude", "Aggregate rating"]
].corr()

location_rating_correlation
```

```python
city_rating_analysis = df.groupby("City").agg(
    Restaurant_Count=("Restaurant ID", "count"),
    Average_Rating=("Aggregate rating", "mean")
)

city_rating_analysis = city_rating_analysis[
    city_rating_analysis["Restaurant_Count"] >= 10
]

top_rated_cities = city_rating_analysis.sort_values(
    by="Average_Rating",
    ascending=False
).head(10)

top_rated_cities
```

```python
top_rated_cities["Average_Rating"].plot(kind="bar")

plt.title("Top Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```
